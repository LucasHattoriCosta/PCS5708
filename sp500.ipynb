```python
import yfinance as yf
import numpy as np
```

```python
original_df = yf.download(tickers="^GSPC", interval="3mo")
```

```python
df = original_df.copy()
df = df.iloc[:-1]
df = df[['Adj Close']].reset_index()
df['month'] = df['Date'].map(lambda x: x.month)
df['year'] = df['Date'].map(lambda x: x.year)
df = df.loc[df['month']==6]
df['Annual Returns'] = np.log(df['Adj Close']) - np.log(df['Adj Close'].shift(1))
df
```

```python
enconding_sp500 = {}
for y in np.arange(2009, 2017+1):
    if df.loc[df.year==y, 'Annual Returns'].iloc[0] >= df.loc[df.year<y, 'Annual Returns'].quantile(0.50):
        enconding_sp500[y] = 1
    else:
        enconding_sp500[y] = 0
```
